```python
import sqlite3

def load_subjects_from_db(db_path):
    # Connect to the SQLite database
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    
    # Dictionary to hold the subjects
    subjects = {}
    
    # Query to select all records from the subjects table
    query = "SELECT Year, Term, Code, Prerequisites, Co_requisites FROM ECE2021"
    
    # Execute the query and fetch all results
    cursor.execute(query)
    rows = cursor.fetchall()
    
    # Process each row
    for row in rows:
        year = row[0]  # Year
        term = row[1]  # Term
        subject_code = row[2]  # Subject Code
        prerequisites = row[3].split(',') if row[3] else []  # Prerequisites
        corequisites = row[4].split(',') if row[4] else []  # Corequisites
        
        # Create a semester identifier
        semester = f"Year {year} - Semester {term}"
        
        # Initialize the semester in the subjects dictionary if not already present
        if semester not in subjects:
            subjects[semester] = {}
        
        # Add the subject with its prerequisites and corequisites
        subjects[semester][subject_code] = {
            "prerequisites": [prereq.strip() for prereq in prerequisites],
            "corequisites": [coreq.strip() for coreq in corequisites]
        }
    
    # Close the database connection
    conn.close()
    
    return subjects

# Example usage
if __name__ == "__main__":
    database_path = 'ece.db'  # Change this to your actual database path
    subjects_dict = load_subjects_from_db(database_path)
    print(subjects_dict)
```
